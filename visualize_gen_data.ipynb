```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
gen_data = pd.read_csv('generated_data/fault_01/gen_data_01.csv', index_col=None)
gen_data = gen_data.drop(columns=['Unnamed: 0'])
print(f"gen_data: {gen_data.shape}")
```

```python
# plot real faulty data to visually evaluating the gen time serires
tep_fault = np.genfromtxt("tep_data/d01_te.dat")

data_fault_forShow_long = tep_fault[200:700]
data_fault_forShow = tep_fault[200:600]

for i in range(tep_fault.shape[1]):
    plt.rcParams['figure.figsize'] = [7, 4]
    plt.figure(i)
    plt.title(i)
    plt.plot( data_fault_forShow_long[:,i],'y')
    plt.plot( np.concatenate((data_fault_forShow[:,i], gen_data.iloc[:,i]), axis = 0),'r' )
    plt.plot( data_fault_forShow[:,i])
```
